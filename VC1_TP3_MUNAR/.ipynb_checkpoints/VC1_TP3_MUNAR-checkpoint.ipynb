```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import os
from PIL import Image
```

```python
# Importo las imagenes a un diccionario.
def obtener_imagenes(ruta):

    # Ruta de la carpeta que contiene las imágenes
    carpeta_imagenes = ruta

    # Obtener la lista de nombres de archivo de las imágenes en la carpeta
    archivos = os.listdir(carpeta_imagenes)

    # Crear un diccionario para almacenar las imágenes
    imagenes_diccionario = {}

    # Iterar a través de los archivos y cargar las imágenes en el diccionario
    for archivo in archivos:
        # Ignorar archivos que no son imágenes
        if archivo.endswith((".jpg", ".jpeg", ".png", ".gif")):
            ruta_completa = os.path.join(carpeta_imagenes, archivo)
            nombre_sin_extension = os.path.splitext(archivo)[0]
            imagen = Image.open(ruta_completa)
            imagenes_diccionario[nombre_sin_extension] = {'img': cv.imread(ruta_completa)}
       
    return imagenes_diccionario

imagenes = obtener_imagenes('./images')

# Obtengo las imagenes en escala de grises y en RGB.
for name, dicc in imagenes.items():
    dicc['img'] = cv.cvtColor(dicc['img'], cv.COLOR_BGR2RGB)
    dicc['img_gray'] = cv.cvtColor(dicc['img'], cv.COLOR_BGR2GRAY)
```

```python
# Ploteo

# Número de filas y columnas para el gráfico
num_filas = len(imagenes)
num_columnas = 1

# Grafico con subplots
fig, axs = plt.subplots(num_filas, num_columnas, figsize=(15, 15))

# Itero sobre el diccionario
for i, (nombre, dicc) in enumerate(imagenes.items()):
    fila = i
    axs[fila].imshow(dicc['img']) 
    axs[fila].set_title(f"{dicc['img'].shape} pix", fontsize=10)        
    
# Ploteo
plt.tight_layout()
plt.show() 
```

```python
# Veamos el template.
template = cv.imread('./template/pattern.png',0)

print('Template')
plt.figure()
plt.imshow(template, cmap='gray')
plt.show()

# Dimensión de la imagen
print(f'Tamaño imagen: {template.shape} pix')
```

```python
# Ajusto el template, si me alcanza el tiempo automatizo este paso

template_ajustado = template[31:150, 11:388]
plt.figure()
plt.imshow(template_ajustado, cmap='gray')
plt.show()

# Dimensión de la imagen
print(f'Tamaño imagen: {template_ajustado.shape} pix')
```

```python
def match_template_unico(imagen, 
                         imagen_gray, 
                         template, 
                         e_img, 
                         e_temp, 
                         metodo):

    # Escalo imagenes
    img_c = cv.resize(imagen_gray, None, fx=e_img, fy=e_img)
    template_c = cv.resize(template_ajustado, None, fx=e_temp, fy=e_temp)

    # Aplico el Detector de Canny
    # Aplico Canny
    img_canny = cv.Canny(img_c, threshold1=50, threshold2=100)
    plt.imshow(img_canny)

    # Aplico el Detector de Canny
    template_canny = cv.Canny(template_c, threshold1=50, threshold2=100)
    plt.imshow(template_canny)

    w, h = template_canny.shape[::-1]
    img_salida = cv.resize(imagen, None, fx=e_img, fy=e_img)
    method = eval(metodo)
    
    # Aplicamos la coincidencia de patrones
    res = cv.matchTemplate(img_canny, template_canny, method)
    
    # Encontramos los valores máximos y mínimos
    min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)
    
    # Si el método es TM_SQDIFF o TM_SQDIFF_NORMED, tomamos el mínimo
    top_left = max_loc
    
    # Marcamos el lugar donde lo haya encontrado
    bottom_right = (top_left[0] + w, top_left[1] + h)
    cv.rectangle(img_salida,top_left, bottom_right, (0,255,0), 2)
    
    # Graficamos el procesamiento y la salida
    #----------------------------------------
    plt.figure()
    
    # Resultado de coincidencia
    plt.subplot(121),plt.imshow(res,cmap = 'gray')
    plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
    
    # Imagen original con recuadros
    plt.subplot(122),plt.imshow(img_salida, cmap = 'gray')
    plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
    
    plt.show()
```

2. Al igual que en el ejercicio 1 realizo una calibración manual por falta de tiempo. Vale mencionar que el algoritmo es muy sensible al tamaño del template y el threshold, lo que lo hace práctico solo en condiciones controladas de distancia al objeto en las imágenes.

```python
# Imágenes
img_gray = imagenes['coca_multi']['img_gray']
img_rgb = imagenes['coca_multi']['img']

# Aplico el detector de Canny, sin este paso el resultado es inaceptable
img_canny = cv.Canny(img_gray, threshold1=50, threshold2=100)

# Modifico el template tambien. Lo escalo a un valor calibrado.
template_canny = cv.Canny(template_ajustado, threshold1=50, threshold2=100)
template_canny = cv.resize(template_canny, None, fx=0.24, fy=0.24)

# Aplico el matchTemplate, calibro un valor de threshold
w, h = template_canny.shape[::-1]
res = cv.matchTemplate(img_canny,template_canny,cv.TM_CCOEFF_NORMED)
threshold = 0.124

loc = np.where( res >= threshold)
for pt in zip(*loc[::-1]):
    cv.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,255,0), 2)

print('Detección múltiple Coca Cola')
plt.figure()
plt.imshow(img_rgb)
plt.show()
```
